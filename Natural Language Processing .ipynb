```
df['text_len'] = df['text'].apply(len)
```

```
#cleaning of text data
import re
def preprocessor(text):
    text = re.sub('<[^>]*>','', text)
    emoticons = re.findall('(7::|;|=)(?:-)?(?:\)|\(|D|P)', text)
    text = re.sub('[\W]+', ' ', text.lower()) + \
                  ' '.join(emoticons).replace('-','')
    return text                     
```

```
df['text'] = df['text'].apply(preprocessor)
```

```
df['text']
```

```
#word stemming, an algorithm for suffix strippng Porter
porter = PorterStemmer()
def tokenizer_porter(text):
    return [porter.stem(word) for word in text.split()]
```

```
df['text'] = df['text'].apply(tokenizer_porter)
```

```
df['text']
```

```
#dealing with stopwords
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize 
from nltk.corpus import brown
import nltk
nltk.download('punkt')
```

```
stop = set(stopwords.words('english')) 
df['text'].apply(lambda x: [item for item in x if item not in stop])
```

```
df['text']=[" ".join(review) for review in df['text'].values]
```

```
#Convert a collection of text documents to a matrix of token counts
class_ = df[(df.stars==1) | (df.stars==5)]
X = class_['text']
y = class_['stars']
cv = CountVectorizer()
X = cv.fit_transform(X)
```

```
print(cv.vocabulary_)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

```
#MultinomialNB and create an instance of the estimator
nb = MultinomialNB()
nb.fit(X_train,y_train)
```

```
#confusion matrix and classification report using these predictions and y_test 
predictions = nb.predict(X_test)
print(confusion_matrix(y_test,predictions))
print('\n')
print(classification_report(y_test,predictions))
```

```
#include TF-IDF to this process using a pipeline
tfidf = TfidfVectorizer(strip_accents=None,lowercase=False,preprocessor=None)
```

```
# pipeline with the following steps:CountVectorizer(), TfidfTransformer(),MultinomialNB()
pipeline = Pipeline([
                     ('bow', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                     ('classifier', MultinomialNB()),

])
```

```
X = class_['text']
y = class_['stars']
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3,random_state=101)
```

```
pipeline.fit(X_train,y_train)
```

```
predictions = pipeline.predict(X_test)
```

```
print(confusion_matrix(y_test, predictions))
print(classification_report(y_test, predictions))
```
