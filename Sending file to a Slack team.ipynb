```
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize 
import json
```

```
get_json=[]
url = 'https://api.hh.ru/vacancies'
par = {'text':'Работа Данными','per_page':'100','period':7,'page':10}
r = requests.get(url, params=par)
```

```
d = json.loads(r.text)
```

```
print(d['items'])
```

```
vacancy = pd.DataFrame(d['items'], columns = ['id','name', 'published_at', 'url','contacts','address', 'area','alternate_url','created_at'])
```
